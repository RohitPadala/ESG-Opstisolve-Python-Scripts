```python
def dump_to_mongo(path):
    file = open(path)
    data = json.load(file)
    client = MongoClient("localhost:27017")
    db = client.esg
    collection = db.esg_report

    try:
        collection.insert_many(data, ordered=False)
        print("inserted ")
    except BulkWriteError as bwe:
        print(bwe.details)
        raise

    client.close()
    print("client closed")
```

```python
for file in files:
    sys_pdf_path = file
    output_json_path = file.split(".")[0] + ".json"
    token = 512
    print(file)
    with open(sys_pdf_path, "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfFileReader(pdf_file)

        final_data = []
        for page_num in range(pdf_reader.numPages):
            try:
                print(page_num)
                page = pdf_reader.getPage(page_num)
                text = page.extractText()
                s_text = summarize_in_gpt(text, token)
                try:
                    json_file = json.loads(s_text)
                    json_file["company"] = "placeholder"
                    json_file["page_number"] = page_num + 1
                    final_data.append(json_file)
                except:
                    try:
                        json.loads(s_text + '"}')
                        json_file["company"] = "placeholder"
                        json_file["page_number"] = page_num + 1
                        final_data.append(json_file)
                    except:
                        continue
                print("processed")
            except:
                print("error")
                continue
        json_data = json.dumps(final_data, indent=4)
        with open(output_json_path, "a") as json_file:
            json_file.write(json_data)
        print("file saved!")
```
